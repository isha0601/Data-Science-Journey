# K-Nearest Neighbors(Classification)

```python
# also called as lazy learner algorithm
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv('Social_Network_Ads.csv')
```

```python
dataset.head(3)
```

```python
# checking distribution of this dataset
sns.scatterplot(x='Age',y='EstimatedSalary',data=dataset,hue='Purchased')
plt.show()
```

```python
dataset.isnull().sum()
```

```python
x=dataset.iloc[:,:-1]
```

```python
y=dataset['Purchased']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
sc.fit(x)
```

```python
sc.transform(x)
```

```python
x=pd.DataFrame(sc.transform(x),columns=x.columns)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# for getting right n_neighbors values
for i in range(1,30):
    knn1=KNeighborsClassifier(n_neighbors=i)
    knn1.fit(x_train,y_train)
    print(i,knn1.score(x_train,y_train)*100, knn1.score(x_test,y_test)*100)
```

```python
knn=KNeighborsClassifier(n_neighbors=3) #default - n_neighbors is 5
# default distance is euclidean(p=2) distance but if we take p=1 then manhattan distance is taken
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)*100
```

```python
knn.score(x_train,y_train)*100
```

```python
knn.predict([[-1.78179743, -1.49004624]])
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=knn)
plt.show()
```

# Hyperparameter Tuning

```python
# Hyperparamter Tuning
# two strategies for hyperparameter tuning
# 1) GridSearchCV  - it is a technique to search through the best paramter values from the given set of the grid of paramters
# 2) RandomizedSearchCV -- 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv('Polynomial.csv')  
```

```python
dataset.head(3)
```

```python
x=dataset.iloc[:,:-1]
```

```python
y=dataset['Salary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
```

```python
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)*100
```

```python
dt.score(x_train,y_train)*100
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
df = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter': ['best', 'random'],
    'max_depth': [i for i in range(2, 20)]
}
```

```python
gd=GridSearchCV(DecisionTreeRegressor(),param_grid=df)
```

```python
gd.fit(x_train,y_train)
```

```python
gd.best_params_
```

```python
gd.best_score_
```

```python
dt=DecisionTreeRegressor(criterion='squared_error',max_depth=2,splitter='best')
```

```python
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)*100
```

```python
dt.score(x_train,y_train)*100
```

```python
df = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter': ['best', 'random'],
    'max_depth': [i for i in range(2, 20)]
}
```

```python
rd= RandomizedSearchCV(DecisionTreeRegressor(),param_distributions=df,n_iter=20)
```

```python
rd.fit(x_train,y_train)
```

```python
rd.best_params_
```

```python
rd.best_score_
```
