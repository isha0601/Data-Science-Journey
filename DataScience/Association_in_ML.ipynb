```python
# Association rule learning 
# Association rule learning is a machine learning techinique used to discover interesting
# relationships pr patterns in large datasets.
# It is often applied to transactional data, where items are bought or used together.

# How does Association rule learning work
# Association rule learning works on the concept of if and else statement, such as if A then B.
# 'If' element is called antecedent, and 'then' statement is called as Consequent.


# How does Association rule learning work
# 1) Support
# 2) Confidence
# 3) Lift
```

```python
# 1) SUPPORT
# The first step for us and the algorithm is to find frequently bought items.
# It is a straightforward calculation that is based on frequency.
# Support(A)= Transactions(A)/(Total Transactions)
```

```python
# 2) CONFIDENCE
# We have identified frequently bought items let`s calculate confidence.
# This will tell us how confident (based on our data) we can be that an item 
# will be purchased, given that another item has been purchased.
# Formula:
# conf(X=>Y) = P(Y|X) = supp(X n Y) /supp(X) = (number of transactions containing X and Y)/(number of transactions containing X)
```

```python
# LIFT
# given that different items are bought at different frequencies(strong association)
# Formula: 
# lift(X=>Y) supp(X n Y)/(supp(X)*supp(Y)) 
# lift>1 : means that the two items are more likely to be bought together
# lift<1 : means that the two items are more likely to be bought separatly
# lift=1 : means that there is no association between the two itmes
```

```python
# Types of Association Rule Learning
# 1) Apriori
# 2) Eclat
# 3) F-P Growth Algorithm
```

# FP- Growth Algorithm

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
pd.set_option('display.max_rows',500)
```

```python
dataset=pd.read_csv('groceries.csv')
```

```python
dataset.head(3)
```

```python
dataset.shape
```

```python
market=[]
for i in range(0,dataset.shape[0]):
    cus=[]
    for j in dataset.columns:
      if type(dataset[j][i])==str:
          cus.append(dataset[j][i])
    market.append(cus)
        
```

```python
l=[]
for i in market:
    for j in i:
        l.append(j)
```

```python
import collections
```

```python
p=collections.Counter(l)
```

```python
d={'Item Name':p.keys(),'values':p.values()}
```

```python
pd.DataFrame(d).sort_values(by=['values'],ascending=False)
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr= TransactionEncoder()
tr.fit(market)
```

```python
tr.transform(market)
```

```python
df=pd.DataFrame(tr.transform(market),columns=tr.columns_)
```

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
fpgrowth(df,min_support=0.07,use_colnames=True,max_len=3).sort_values(by=['support'])
```
