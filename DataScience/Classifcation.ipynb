# Confusion Matrix / Error Matrix

```python
# formula 
# Accuracy= (TP+TN)/(TP+TN+FP+FN)
# Error=(FP+FN)/(TP+TN+FP+FN) = 1-Accuracy
# Recall = TP/(TP+FN)  
#precision=TP/(TP+FP)
```

```python
#f1 score  is the harmonic mean of precision and recall 
# f1 score= 2* [(precision * recall)/(precision + recall)]
# jiska  jyda f1 wahi model select karenge 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('placement_1.csv')
```

```python
dataset.head(3)
```

```python
x=dataset.iloc[:,:-1]
```

```python
y=dataset['placed']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
#creating confusion matrix
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score
```

```python
cf=confusion_matrix(y_test,lr.predict(x_test))
cf
```

```python
sns.heatmap(cf,annot=True)
plt.show()
```

```python
precision_score(y_test,lr.predict(x_test))*100
```

```python
recall_score(y_test,lr.predict(x_test))*100
```

```python
f1_score(y_test,lr.predict(x_test))*100
```

# Naive Bayes Algorithm

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("placement_1.csv")
```

```python
dataset.head(3)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
dataset.isnull().sum()
```

```python
sns.kdeplot(data=dataset['cgpa'])  # to check if it is normally distributed or not
plt.show()
```

```python
sns.kdeplot(data=dataset['score'])  # to check if it is normally distributed or not
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x='cgpa',y='score',data=dataset,hue='placed')
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset['placed']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100
```

```python
gnb.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)*100
```

```python
mnb.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```

```python
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)*100
```

```python
bnb.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```
