```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
# y=m1x1+m2x2^2+c
#y=-137634.60994912*x1 + 20502.04918033*x2^2 + 231288.51045789
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
prd=lr.predict(x)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv('Housing.csv')
```

```python
dataset.head(3)
```

```python
dataset = dataset.select_dtypes(include=['number'])
# Compute and plot correlation matrix
sns.heatmap(data=dataset.corr(), annot=True, cmap='coolwarm')  # optional: change color scheme
plt.show()
```

```python
x=dataset.iloc[:,:-1]
```

```python
x
```

```python
y=dataset['price']
```

```python
sc=StandardScaler()
sc.fit(x)
```

```python
x=pd.DataFrame(sc.transform(x),columns=x.columns)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression ,Lasso,Ridge
from sklearn.metrics import mean_absolute_error,mean_squared_error
import numpy as np
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
mean_squared_error(y_test,lr.predict(x_test)), mean_absolute_error(y_test,lr.predict(x_test)),np.sqrt(mean_squared_error(y_test,lr.predict(x_test)))
```

```python
lr.coef_
```

```python
plt.bar(x.columns,lr.coef_)
plt.show()
```

```python
plt.bar(x.columns,lr.coef_)
plt.title('LinearRegression')
plt.xlabel('columns')
plt.ylabel('coef')
plt.show()
```

```python


#Lasso /L1 
```

```python
la=Lasso(alpha=0.01)
la.fit(x_train,y_train)
```

```python
la.score(x_test,y_test)*100
```

```python
mean_squared_error(y_test,la.predict(x_test)), mean_absolute_error(y_test,la.predict(x_test)),np.sqrt(mean_squared_error(y_test,la.predict(x_test)))
```

```python
mean_squared_error(y_test,lr.predict(x_test)), mean_absolute_error(y_test,lr.predict(x_test))
```

```python
plt.bar(x.columns,la.coef_)
plt.title('Lasso')
plt.xlabel('columns')
plt.ylabel('coef')
plt.show()
```

```python
#Ridge
```

```python
ri=Ridge(alpha=10)
ri.fit(x_train,y_train)
```

```python
ri.score(x_test,y_test)*100
```

```python
mean_squared_error(y_test,ri.predict(x_test)), mean_absolute_error(y_test,ri.predict(x_test)),np.sqrt(mean_squared_error(y_test,ri.predict(x_test)))
```

```python
plt.bar(x.columns,ri.coef_)
plt.title('Lasso')
plt.xlabel('columns')
plt.ylabel('coef')
plt.show()
```
